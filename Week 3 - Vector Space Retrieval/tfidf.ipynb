# TF\*IDF Assignment

Go to [this](www.cmry.nl/tfidf) webpage, and enter your u-number (without the u). **Be sure you've typed it correctly, it determines your data!** You are provided with several documents, and a label: 9 for training, 1 for testing (indicated by the `?` label). Your assignment is to caculate TF\*IDF weights for each of their word vectors, normalize them, and predict the label for the closest document to the test document.

## Grading

There are 4 grading components to this assignment: (1) correctly constructing a document \* term frequency matrix, (2) succesfully weighting them with TF\*IDF, (3) normalization, (4) predicting the correct label. All assigments will be checked by hand, so it's important that you write out as much of the formulas as you can. As such, if you make a mistake in the beginning, I can still allocate points.

## How to Report

To make checking a bit more easy though, please follow these instructions. Work in a .txt file in a plain text editor of your choice: e.g. notepad (windows), texteditor (mac), gedit (linux), notepad++, vi(m), emacs, nano, sublime text, etc. Write out the formulas as you would code (e.g. sqrt for square root, ^2 for squared, etc). For the tables, you can use markdown formatting, like so:
