### Unsupervised Topic Models

- Topic Models are in general unsupervised.
- Given some $X$ amount of topics (same with $k$-means amount of clusters), divide the data amongst these.
- Learning mechanism is [complex](https://en.wikipedia.org/wiki/Latent_Dirichlet_allocation#Inference), a very simplified version can be found [here](http://blog.echen.me/2011/08/22/introduction-to-latent-dirichlet-allocation/).
- They are *generative* Bag of Words models; so we can ask to generate a document on `topic 2` (sports) and `4` (fruit), and it might generate the following:

`from the goal shot banana juice` 

### General Intuition

- Words that co-occur a lot in texts probably relate to the same topic.
- Words that occur a lot in general are not informative to these topics.
- Given a set of documents, topics themselves have some probability of occuring.
- Given a set of topics, words that are very characteristic for some topic should have a high probability.

### Learning Under Supervision

- Labels are given.
- Train some Topic Model (Latent Dirichlet Allocation by David Blei is a popular example).
- Output are all topics assigned some probability of being in the document. For example:
    - Actual labels: `sports, finance`.
    - Predicted labels: `sports: 0.2, finance: 0.3, climate: 0.01, united nations: 0.05, AI: 0.02, transport: 0.1`.
- Evaluation is not standard (i.e. not just accuracy): we want to know if the labels of our test document are top ranked in the output.

### Set-up

- Label all comments with the label of the article they were posted under.
- We can have comments that are actually on another topic than the article.
- While training, we don't care, and assume these are a minority.
- We evaluate the model by testing if they predict the articles well.

### Application

- Because LDA doesn't pick up on literal messages, we can apply it to the full data again.
- Comments will now actually be labelled differently then they were while training (because some other topics are more salient).
- We hope to correctly re-label comments to determine if they are on-topic (same as article), or off-topic.
